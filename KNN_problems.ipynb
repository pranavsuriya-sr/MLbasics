QUESTION - 1 - 10,6

```python
import pandas as pd
import numpy as np
import math
```

```python
db = pd.read_csv("/content/ques1.csv")
db
```

```python
first_row_voltage = db.iloc[0]['Voltage']
print("Voltage in the first row:", first_row_voltage)
```

```python
for i in range(0, 10):
    vol = db.iloc[i]['Voltage']
    cur = db.iloc[i]['Current']
    db.loc[i, "distance"] = np.sqrt((10 - vol)**2 + (6 - cur)**2)
    print(db.loc[i, "distance"])
db
```

```python
db['rank'] = db['distance'].rank()
db['rank'] = db['rank'].astype(int)
print(db)
```

```python
rank_1_condition = db[db['rank'] == 1]['Condition'].mode().values[0]
rank_2_condition = db[db['rank'] == 2]['Condition'].mode().values[0]
rank_3_condition = db[db['rank'] == 3]['Condition'].mode().values[0]

# Determine the majority condition among ranks 1, 2, and 3
conditions = [rank_1_condition, rank_2_condition, rank_3_condition]
majority_condition = max(conditions, key=conditions.count)

# Output the result
print("The majority condition for ranks 1, 2, and 3 is:", majority_condition)
```
