Question - 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
db = pd.read_csv("/content/ques1.csv")
db
```

```python
db = pd.read_csv("/content/ques1.csv")
db.columns = ['Marks', 'Percentile']
model = linear_model.LinearRegression()
model.fit(db[['Marks']], db['Percentile'])
print(model.predict([[115]]))
print("m = ",model.coef_)
print("b = ", model.intercept_)
```
